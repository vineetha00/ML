```python
import seaborn as sns
from sklearn import preprocessing
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
x = np.array([1,2,3,4])
y = np.array([5,6,7,8])
```

Gradient Descent for Linear Regression

```python
def grad_desc(x,y):
    w_curr = 0
    b_curr = 0
    m = len(x)
    iter = 1000
    a = 0.1
    for i in range(iter):
        y_pred = w_curr*x + b_curr
        dw = 2/m*sum((y_pred-y)*x)
        db = 2/m*sum((y_pred-y))
        w_curr = w_curr - a*dw
        b_curr = b_curr - a*db
        cost_fn = (1/(2*m)) * sum((y-y_pred)**2)
        print(f"{w_curr} {b_curr} {i+1} {cost_fn}")
```

```python
grad_desc(x,y)
```

Gradient Descent for Logistic Regression

```python
def grad_desc1(x,y):
    w_curr = 0
    b_curr = 0
    m = len(x)
    iter = 1000
    a = 0.002
    for i in range(iter):
        z = np.dot(x, w_curr) + b_curr
        z = np.clip(z, -500, 500)  # Clip z to avoid overflow
        y_pred = 1 / (1 + np.exp(-z))
        dw = 1/m*(np.dot(x.T,(y_pred-y)))
        db = 1/m*np.sum((y_pred-y))
        w_curr = w_curr - a*dw
        b_curr = b_curr - a*db
        y_pred = np.clip(y_pred, 1e-10, 1 - 1e-10)  # to avoid log(0)
        cost_fn = -1/m * np.sum(y*np.log(y_pred)+(1-y)*np.log(1-y_pred))
        print(f"{w_curr} {b_curr} {i+1} {cost_fn}")
```

```python
x1 = np.array([22,55,77,99,11])
y1 = np.array([0,0,1,0,1])
grad_desc1(x1,y1)
```
