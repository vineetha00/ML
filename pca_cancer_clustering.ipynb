```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import StandardScaler

scaler = StandardScaler()
df_std = scaler.fit_transform(df)

pca = PCA(n_components=9)
pca.fit(df_std)

df_pca = pca.transform(df_std)

explained_variance_ratio = pca.explained_variance_ratio_

print(explained_variance_ratio)
print(df_pca)
```

```python
df1 = pd.DataFrame(df_pca)
df1.head()
```

```python
for i in df1.columns:
    sns.boxplot(df1[i])
    plt.show()
```

```python
df1.columns
```

```python
def iqr(df,x):
    q1 = np.percentile(x,25)
    q3 = np.percentile(x,75)
    i = q3-q1
    lf = q1 - 1.5*i
    hf = q3 + 1.5*i
    df = df[(x >= lf) & (x <= hf)]
    return df
df1 = iqr(df1,df1[0])
df1 = iqr(df1,df1[1])
df1 = iqr(df1,df1[2])
df1 = iqr(df1,df1[3])
df1 = iqr(df1,df1[4])
df1 = iqr(df1,df1[6])
df1 = iqr(df1,df1[7])
df1 = iqr(df1,df1[8])
```

```python
corr = df1.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True)
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=2)
y_predicted= km.fit_predict(df1[[1,4]])
y_predicted
```

```python
df1['clusters']= y_predicted
```

```python
sns.scatterplot(x=df1[1], y= df1[4], hue = df1['clusters'])
```
