```python
import seaborn as sns
from sklearn import preprocessing
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df = np.array([[0.8,0.44],[-1.28,-1.43],[0.49,0.99]])
```

```python
#Standardize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

#Apply PCA without specifying the number of components
pca = PCA()  # No n_components specified, so it will keep all components
pca.fit(scaled_data)

#Check explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = explained_variance_ratio.cumsum()

for i, variance in enumerate(cumulative_variance):
    print(f"Component {i+1}: Cumulative Variance = {variance:.2f}")

# Plot cumulative variance to decide the number of components
import matplotlib.pyplot as plt

plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA - Cumulative Explained Variance')
plt.grid(True)
plt.show()
```
